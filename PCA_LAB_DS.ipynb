```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
```

```python
df1=pd.read_csv('/content/survey lung cancer.csv')
df1['LUNG_CANCER'] = df1['LUNG_CANCER'].replace({'YES': 1, 'NO': 0})
print(df1.head)
```

```python
df2=pd.read_csv('/content/Prostate_Cancer.csv')
df2.drop('diagnosis_result', axis=1, inplace=True)
print(df2.head)
```

```python
df3=pd.read_csv('/content/kag_risk_factors_cervical_cancer.csv')
df3.replace({'?': 0}, inplace=True)
print(df3.head)
```

```python
df4=pd.read_csv('/content/Cancer_Rates.csv')
print(df4.head)
```

```python
df5=pd.read_csv('/content/breastCancer.csv')
df5.replace({'?': 0}, inplace=True)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
print(df5.shape)
```

```python
scaler = StandardScaler()
x1 = df1.iloc[:,1:]
y1 = df1.iloc[:,0]
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1,y1,test_size=0.2,random_state=42)
x_train1 =scaler.fit_transform(x_train1)
x_test1 = scaler.transform(x_test1)
pca_model = PCA(n_components=11)
x_train_trf1 = pca_model.fit_transform(x_train1)
x_test_trf1 = pca_model.fit_transform(x_test1)
x_train_trf1.shape
```

```python
plt.plot(np.cumsum(pca_model.explained_variance_ratio_))
```
